# group facility_type

```python
climatedata[['Year_Factor','State_Factor','building_class','facility_type']] = climatedata[['Year_Factor','State_Factor','building_class','facility_type']].astype('category')
facility = ['Commercial','Education','Food','Health','Industrial','Lodging','Multifamily','Office','Public_Assembly','Public_Safety','Retail','Service','Warehouse']
facility1 = ['Parking','Laboratory','Nursing','Building','Data','Religious']
categorytype = ['Commercial','Education','Food','Health','Industry','Lodging','Multifamily','Office','Public_area','Public_safety','Retail','Service','Warehouse']
categorytype1 = ['Commercial','Health','Lodging','Multifamily','Office','Public_area']
for i, j in zip(facility,categorytype):
    climatedata.loc[climatedata['facility_type'].str.contains(i,case=False), 'facility_type_1'] = j
for i1, j1 in zip(facility1,categorytype1):
    climatedata.loc[climatedata['facility_type'].str.contains(i1,case=False), 'facility_type_1'] = j1
climatedata.fillna({'facility_type_1':'Multifamily'},inplace=True)
climatedata.drop('facility_type',axis=1,inplace=True)
climatedata.rename(columns={'facility_type_1':'facility_type'},inplace=True)
```

# Get ready for Scaling and encoding

```python
categories = x_train.select_dtypes('category').columns.tolist()
numerical = x_train.select_dtypes('number').columns.tolist()
```

```python
for i in ['direction_max_wind_speed','direction_peak_wind_speed','max_wind_speed']:
    numerical.remove(i)
```

```python
numerical
```

```python
simplenumerical = ['direction_max_wind_speed','direction_peak_wind_speed','max_wind_speed']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector as selector
```

```python
simple_pipe = Pipeline(steps=[("simpleimpute", SimpleImputer(missing_values=np.nan,strategy="most_frequent")),("scale",StandardScaler())])
knn_pipe = Pipeline(steps=[("knnimpute", KNNImputer(n_neighbors=5)),("scale",StandardScaler())])
cat_pipe = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(transformers=[("nums",simple_pipe,simplenumerical),
                                                ("numk",knn_pipe,numerical)
                                                ])

preprocessor.fit(x_train)
```

```python
preprocessor_cat = ColumnTransformer(transformers=[("cat",cat_pipe,categories)
                                                ])
preprocessor_cat.fit(x_train)
```

```python
cat_column = (preprocessor_cat.get_feature_names_out()).tolist()
```

```python
simplenumerical.extend(numerical)
```

```python
from sklearn import set_config

set_config(display="diagram")
preprocessor
```

```python
def transform(df,df1,df2):
    num_transform = df.transform(df2)
    num_processed = pd.DataFrame(num_transform,columns = simplenumerical)
    cat_transform = df1.transform(df2)
    cat_processed = pd.DataFrame(cat_transform.toarray(),columns = cat_column)
    baseline_processed = pd.concat([num_processed,cat_processed],axis=1)
    return baseline_processed
```

```python
train_data = transform(preprocessor,preprocessor_cat,x_train)
valid_data = transform(preprocessor,preprocessor_cat,x_valid)
test_data = transform(preprocessor,preprocessor_cat,x_valid)
```

```python
train_data.to_parquet('train_processed.parquet')
valid_data.to_parquet('valid_processed.parquet')
test_data.to_parquet('test_processed.parquet')
```
